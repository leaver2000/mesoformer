```python
import dataclasses
from mesoscaler._typing import Pair
from typing import NamedTuple


class Shape(NamedTuple):
    c: int
    t: int
    z: int
    y: int
    x: int


@dataclasses.dataclass(
    # kw_only=True,
)
class Config:
    dx: int
    dy: int
    dimensions: Shape
    patch: Shape
    # width: int
    # height: int
    # dx: int
    # dy: int
    # ratio: float

    # time_n: int = 1
    # time_patch_size: int = 1

    # @classmethod
    # def from_ratio(cls, img_size: int, distance: int, ratio: float):
    #     height = int(img_size / ratio)
    #     width = img_size
    #     dx = distance
    #     dy = int(dx / ratio)
    #     return cls(
    #         width=width,
    #         height=height,
    #         dx=dx,
    #         dy=dy,
    #         ratio=ratio,
    #     )

    # @property
    # def image_size(self) -> Pair[int]:
    #     return (self.width, self.height)

    # @property
    # def patch_size(self) -> Pair[int]:
    #     return (self.dx, self.dy)

    def _validate(self):
        for x, y in zip(self.dimensions, self.patch):
            assert x % y == 0
    # @property
    # def patch_size(self) -> tuple[int,int]:
    #     return self.patch.y, self.patch.x

    # @property
    # def image_size(self) -> tuple[int,int]:
    #     return self.dimensions.y, self.dimensions.x
    # #     img_size = self.image_size
    # #     patch_size = self.patch_size
    # #     frames = self.time_n
    # #     t_patch_size = self.time_n
    # #     assert img_size[1] % patch_size[1] == 0
    # #     assert img_size[0] % patch_size[0] == 0
    # #     assert frames % t_patch_size == 0


    @classmethod
    def from_ratios(
        self,
        dimensions: tuple[int,int,int,int,int],
        d_ratio:float = 2.5,
        xy_ratio:float = .2,
    ):
        c, t, z, y, x = dimensions
        return self(
        dx=int(d_ratio  * x),
        dy=int(d_ratio  * y),
        dimensions=Shape(
            c,
            t,
            z,
            y,
            x,

        ),
        patch=Shape(
            c,
            t,
            z,
            int(y * xy_ratio),
            int(x * xy_ratio),
        ),

        
    )
    # @property
    # def frames(self) -> int:
    #     return self.dimensions.t
    # @property
    # def patch_size
    def to_dict(self):
        c, t, z, y, x = self.dimensions
        pc, pt, pz, py, px = self.patch
        return {
            "image_size": (y, x),
            "patch_size": (py, px),
            "in_channels": c,
        }

config = Config.from_ratios(
    (5, 12, 10, 40, 80),
    d_ratio=2.5,
    xy_ratio=.2,
)
config._validate()
config.to_dict()
```
