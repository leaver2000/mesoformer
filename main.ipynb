```python
def lonlats2xy(
    crs: pyproj.CRS, lons: NDArray[np.float_], lats: NDArray[np.float_], return_extent: bool = False
) -> tuple[NDArray[np.float_], ...]:
    proj = pyproj.Proj(crs)
    xxyy = np.stack(proj.transform(lons, lats, radians=False), axis=-1)
    x, y = np.median(xxyy[..., 0], axis=0), np.median(xxyy[..., 1], axis=1)
    if return_extent:
        return x, y, np.array([x.min(), y.min(), x.max(), y.max()])
    return x, y


def _prepare_era5_dataset(ds: xr.Dataset) -> xr.Dataset:
    raise NotImplementedError


CoordinateMap = Iterable[tuple[cf.StandardName, NDArray[np.float_]]]


def create_dimension_coordinates(data_coords: CoordinateMap) -> dict[str, xr.DataArray]:
    return {
        coord.name: xr.DataArray(data, name=coord.name, dims=coord.dims, attrs=coord.attrs)
        for coord, data in data_coords
    }


def _prepare_urma_dataset(ds: xr.Dataset, meta: DatasetMetadata) -> xr.Dataset:
    lons, lats = (ds[x].to_numpy() for x in (cf.lon, cf.lat))
    x, y, area_extent = lonlats2xy(meta.crs, lons, lats, return_extent=True)
    area_def = pyresample.geometry.AreaDefinition(
        meta.title, meta.comment, "", meta.crs, x.shape[0], y.shape[0], area_extent=area_extent
    )
    coords = create_dimension_coordinates([(cf.lon, lons), (cf.lat, lats), (cf.x, x), (cf.y, y)])

    return ds.assign_coords(coords).assign_attrs({"area_definition": area_def})


def open_zarr(path: str, meta: DatasetMetadata | None = None) -> xr.Dataset:
    ds = xr.open_zarr(path)
    if meta is not None:
        ds = _prepare_urma_dataset(ds, meta)
    return ds


urma = open_zarr("tests/data/urma.zarr", URMAEnum.md)
era5 = open_zarr("/mnt/data/era5/1959-2022-6h-128x64_equiangular_with_poles_conservative.zarr/")
urma, era5
```

```python
def get_metadata(ds: xr.Dataset, dataset: str, drop=[]):
    return (
        pd.DataFrame(
            [
                ds[v].attrs
                | {
                    "dataset": dataset,
                    "standard_name": v,
                    "dims": ds[v].dims,
                    "coordinates": ds[v].coords.keys(),
                }
                for v in ds.data_vars
            ]
        )
        .drop(columns=drop)
        .to_dict(orient="records")
    )


# {"urma": get_metadata(urma, drop=["grib_id"]), "era5": get_metadata(era5)}


pd.DataFrame(get_metadata(urma, "urma", drop=["grib_id"]) + get_metadata(era5, "era5"))
```
