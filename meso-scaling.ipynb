```python
# arraytools.py
import functools
from typing import Literal, Callable, TypeVar, ParamSpec, Concatenate, Any

from sklearn import preprocessing
from numpy.typing import NDArray, ArrayLike
import torch


FloatingDTypeT = TypeVar("FloatingDTypeT", bound=np.floating, covariant=True)
AnyT_co = TypeVar("AnyT_co", bound=Any, covariant=True)
R_co = TypeVar("R_co", bound=Callable)
P = ParamSpec("P")


def as_any_array(dtype: type[FloatingDTypeT]):
    def decorator(
        func: Callable[Concatenate[NDArray[np.number], P], AnyT_co]
    ) -> Callable[Concatenate[ArrayLike, P], AnyT_co]:
        @functools.wraps(func)
        def wrapper(x: ArrayLike, *args: P.args, **kwargs: P.kwargs) -> AnyT_co:
            return func(np.asanyarray(x, dtype=dtype), *args, **kwargs)

        return wrapper

    return decorator


@as_any_array(np.float32)
def low_high_scaler(x: NDArray[np.number], l: float = 0, h: float = 1.2, *, axis=None) -> NDArray[np.float32]:
    return (x - x.min(axis)) / (x.max(axis) - x.min(axis)) * (h - l) + l


@as_any_array(np.float32)
def min_max_scaler(x: NDArray[np.number], *, axis=None) -> NDArray[np.float32]:
    return (x - x.min(axis)) / (x.max(axis) - x.min(axis))


@as_any_array(np.float32)
def std_scaler(x: NDArray[np.number], *, axis=None) -> NDArray[np.float_]:
    return (x - x.mean(axis)) / x.std(axis)


@as_any_array(np.float32)
def max_abs_scaler(x: NDArray[np.number], *, axis=None) -> NDArray[np.float_]:
    return x / np.abs(x).max(axis)


@as_any_array(np.float32)
def normalize(x: NDArray[np.number]) -> NDArray[np.float_]:
    return (x - x.min(keepdims=True)) / (x.max(keepdims=True) - x.min(keepdims=True))





def _representation(x: Any) -> str:
    if isinstance(x, slice):
        return f"{x.start or ''}:{x.stop or ''}:{x.step or ''}"
    elif x is ...:
        return "..."
    elif x is None:
        return "newaxis"
    elif dtype := getattr(x, "dtype", None):
        return f"{x.__class__.__name__}[{dtype}]"
    return str(x)


def mask_representation(mask: Any) -> str:
    if isinstance(mask, tuple):
        return f"[{', '.join([_representation(m) for m in mask])}]"

    return f"[{_representation(mask)}]"


class MaskedScaler:
    def __init__(
        self,
        data: ArrayLike,
        mask: bool | list[bool] | None | tuple[NDArray[np.bool_], Any | None] = ...,
        low=0.0,
        high=1.0,
        axis=None,
    ) -> None:
        self.data = np.asanyarray(data)
        self._mask = mask
        self.low = low
        self.high = high
        self.axis = axis

    @property
    def mask(self):
        mask = self._mask
        if (mask is ...) or (mask is None) or isinstance(mask, tuple):
            return mask
        return np.asanyarray(mask)

    @property
    def low_high(self) -> NDArray[np.float_]:
        return low_high_scaler(self.data, l=self.low, h=self.high)[self.mask]

    @property
    def min_max(self) -> NDArray[np.float_]:
        return min_max_scaler(self.data)[self.mask]

    @property
    def std(self) -> NDArray[np.float_]:
        return std_scaler(self.data)[self.mask]

    @property
    def max_abs(self) -> NDArray[np.float_]:
        return max_abs_scaler(self.data)[self.mask]

    @property
    def norm(self) -> NDArray[np.float_]:
        return normalize(self.data)[self.mask]

    def to_numpy(self) -> NDArray[np.float_]:
        return np.stack([self.low_high, self.min_max, self.std, self.max_abs, self.norm])

    @property
    def columns(self):
        return ["low_high", "min_max", "std", "max_abs", "norm"]

    def __repr__(self) -> str:
        mask_repr = mask_representation(self.mask)
        data = "\n".join([f"{k}:\n{v}" for k, v in zip(self.columns, self.to_numpy())])
        return f"{self.__class__.__name__}(\n{data}\n){mask_repr}"

    def __getitem__(self, mask):
        self._mask = mask
        return self

    def to_pandas(self):
        df = pd.DataFrame(self.to_numpy().reshape(-1, 5), columns=self.columns)
        return df


scaler = MaskedScaler([[-11, 5], [2, 4], [3, 8]])
scaler
```

```python
import itertools

offsets = [
    (-1,-1),
    (-1,1),
    (1,-1),
    (1,1), 
]
def get_offset(i):
    return offsets[i % len(offsets)]

[(x, get_offset(i))  for i, x in enumerate([
            (0.75, "r"),
            (1.0, "g"),
            (1.225, "b"),
            (1.5, "c"),
            (1.75, "m"),
            (2.0, "y"),
    ]) 
]
from typing import TypeVar, Iterable
T1 = TypeVar("T1")
T2 = TypeVar("T2")

def zip_products(
    *items:T1, products:list[T2] ,
) -> Iterable[tuple[T1,T2]]:
    return zip(items, itertools.cycle(products))

list(zip_products(
    (0.75, "r"),
    (1.0, "g"),
    (1.225, "b"),
    (1.5, "c"),
    (1.75, "m"),
    (2.0, "y"),
    products= [(-1,-1),(-1,1),(1,-1),(1,1)]
)
)
```
