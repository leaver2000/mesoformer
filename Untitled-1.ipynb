```python
WGS84 = pyproj.Geod(ellps="WGS84")


def get_distances(lons: np.ndarray, lats: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    ew = WGS84.inv(lons[:, :-1], lats[:, 1:], lons[:, 1:], lats[:, 1:])[-1]
    ns = WGS84.inv(lons[1:, :], lats[1:, :], lons[1:, :], lats[:-1, :])[-1]
    return ew, ns


def get_squared_area(lons: np.ndarray, lats: np.ndarray) -> np.ndarray:
    ew, ns = get_distances(lons, lats)
    return ew[1:, :] * ns[:, 1:]




# md = URMAEnum.md
assign(ds, URMAEnum.md)
```

```python
from typing import Iterable, TypeVar, Iterable, Callable
import itertools

T1 = TypeVar("T1")
T2 = TypeVar("T2")


def squish(func: Callable[[T1], T2], __iterable: T1 | Iterable[T1], *args: T1) -> list[T2]:
    """
    >>> assert squish(lambda x: x, "foo", "bar", "baz") == ["foo", "bar", "baz"]
    >>> assert squish(lambda x: str(x), range(3), 4, 5) == ["0", "1", "2", "4", "5"]
    >>> assert squish("hello {}".format, (x for x in ("foo", "bar")), "spam") == ["hello foo", "hello bar", "hello spam"]
    """
    it1 = iter([__iterable] if isinstance(__iterable, str) or not isinstance(__iterable, Iterable) else __iterable)
    return [func(x) for x in itertools.chain(it1, args)]  # type: ignore


assert squish(lambda x: x, "foo", "bar", "baz") == ["foo", "bar", "baz"]
assert squish(lambda x: str(x), range(3), 4, 5) == ["0", "1", "2", "4", "5"]
assert squish("hello {}".format, (x for x in ("foo", "bar")), "spam") == ["hello foo", "hello bar", "hello spam"]
```

```python
import enum
import functools
from typing import Literal, TypeVar, Any,Mapping
import datetime

from src.mesoformer.utils import indent_kv
from src.mesoformer.datasets.urma.constants import (
    TOTAL_CLOUD_COVER,
    VISIBILITY,
    CEILING,
    U_WIND_COMPONENT_10M,
    V_WIND_COMPONENT_10M,
    TEMPERATURE_2M,
)


TimeIndexer = slice | list[str | datetime.datetime] | pd.Index | np.ndarray

T = TypeVar("T")

def left_right_both(left: T | None = None, right: T | None = None) -> tuple[T, T]:
    left = left if left is not None else right
    right = right if right is not None else left
    assert left is not None and right is not None
    return left, right


class CFGrids(str, enum.Enum):
    ALBERS = "albers_conical_equal_area"
    AE = "azimuthal_equidistant"
    GEOS = "geostationary"
    LAEA = "lambert_azimuthal_equal_area"
    LCC = "lambert_conformal_conic"
    LCEA = "lambert_cylindrical_equal_area"
    MERC = "mercator"
    PS = "polar_stereographic"
    STEREO = "stereographic"
    TM = "transverse_mercator"

    @classmethod
    def _missing_(cls, value: Any) -> Any:
        return cls._member_map_[str(value).upper()]

CFGrids("laea")
```

```python
def div(*args: Any, **kwargs: Any) -> str:
    return f"""<div {' '.join(f'{k}="{v}"' for k, v in kwargs.items())}>{''.join(args)}</div>"""


def stack_html(ds: xr.Dataset, other: Any) -> str:
    footer = (
        f'<pre>{textwrap.indent(other, "    ")}</pre>'
        if isinstance(other, str)
        else f'<div class="xr-var-data"><pre>{str(other)}</pre></div>'
    )
    html = ds._repr_html_()

    return div(div(html, style="display: inline-block;"), div(footer, style="display: inline-block;"))


MesoScale = Literal["mesoscale-alpha", "mesoscale-beta", "mesoscale-gamma"]


_meso_scale = {"mesoscale-alpha": 2_000, "mesoscale-beta": 200, "mesoscale-gamma": 20}
"""Mesoscale Meteorology is divided into these subclasses:[1]

Meso-alpha 200-2000 km scale of phenomena like fronts, squall lines, mesoscale convective systems (MCS), tropical cyclones at the edge of synoptic scale
Meso-beta 20-200 km scale of phenomena like sea breezes, lake effect snow storms
Meso-gamma 2-20 km scale of phenomena like thunderstorm convection, complex terrain flows (at the edge to microscale, also known as storm-scale)
As a note, tropical and subtropical cyclones are classified by National Hurricane Center as synoptic scale rather than mesoscale.[2]
"""


def _ready_stack(ds: xr.Dataset) -> xr.DataArray:
    # TODO: need to run this on the weatherbench dataset.  This does not account for the level dimension
    if not (
        LONGITUDE in ds
        or LATITUDE in ds
        or TIME in ds
        or {Y_AXIS, X_AXIS, TIME} == set(ds.dims.keys())
        or {LONGITUDE, LATITUDE, TIME} == set(ds.coords.keys())
    ):
        raise ValueError("Dataset must have longitude, latitude, and time coordinates")
    da = ds.set_coords([LONGITUDE, LATITUDE, TIME]).transpose("y", "x", ...).to_array().stack(channel=(TIME, VARIABLE))
    return da



class PartialTargetArea:
    def __init__(
        self,
        area_extent: list[float],
        width: float = 256,
        height: float = 256,
        units="m",
        target_grid: CFGrids | str = "lambert_azimuthal_equal_area",
    ) -> None:
        target_grid = CFGrids(target_grid)
        self.title = title = target_grid.value.replace("_", " ").title()

        self._in_cf = {"grid_mapping_name": target_grid.value, "units": units}
        self._partial = functools.partial(
            pyresample.geometry.AreaDefinition,
            self.__class__.__name__,
            title,
            None,
            height=height,
            width=width,
            area_extent=area_extent,
        )

    @classmethod
    def create(
        cls,
        meso: MesoScale | None = None,
        height: float | None = None,
        width: float | None = None,
        *,
        x: float | None = None,
        y: float | None = None,
        scale_x: float = 1,
        scale_y: float = 1,
        units: str = "m",
        target_grid="lambert_azimuthal_equal_area",
    ) -> "PartialTargetArea":
        kwargs = cls.meter_scale(height, width, meso, x, y, scale_x, scale_y, units)
        kwargs["target_grid"] = target_grid
        return cls(**kwargs)

    @staticmethod
    def meter_scale(
        height: float | None = None,
        width: float | None = None,
        meso: MesoScale | None = None,
        x: float | None = None,
        y: float | None = None,
        scale_x: float = 1,
        scale_y: float = 1,
        units: str = "m",
    ) -> dict[str, Any]:
        if meso is None:
            x, y = left_right_both(x, y)
        else:
            x = y = _meso_scale[meso]
            units = "km"

        height, width = left_right_both(height, width)
        assert units in ("m", "km")

        if units == "km":
            x *= 1000
            y *= 1000

        x /= 2
        y /= 2

        return {
            "area_extent": [-x * scale_x, -y * scale_y, x * scale_x, y * scale_y],
            "width": width * scale_x,
            "height": height * scale_y,
            "units": "m",
        }

    def __call__(self, latitude: float, longitude: float) -> pyresample.geometry.AreaDefinition:
        in_cf = self._in_cf | {"latitude_of_projection_origin": latitude, "longitude_of_projection_origin": longitude}
        return self._partial(projection=pyproj.CRS.from_cf(in_cf))

    def __repr__(self):
        kwargs = "\n".join(indent_kv(*self._partial.keywords.items()))
        return f"{self.__class__.__name__}:\n{self.title}\n{kwargs}"


class GriddedDataSampler(pyresample.geometry.GridDefinition):
    def __init__(self, ds: xr.Dataset, channels: list[CFDatasetEnum], target: PartialTargetArea) -> None:
        super().__init__(lons=(ds[LONGITUDE].to_numpy() - 180) % 360 - 180, lats=ds[LATITUDE].to_numpy())

        enum_set = {type(c) for c in channels}
        if len(enum_set) > 1:
            raise ValueError(f"Channels must be of the same type. Received {enum_set}")
        (enum_cls,) = enum_set

        self._md = enum_cls.md
        self._data = ds[channels].set_coords([LONGITUDE, LATITUDE, TIME])
        self._target = target

    @property
    def time(self) -> xr.DataArray:
        return self._data[TIME]

    @property
    def dims(self):
        return self._data.dims

    def resample_nearest(self, data: np.ndarray, target: pyresample.geometry.AreaDefinition) -> np.ndarray:
        return pyresample.kd_tree.resample_nearest(self, data, target, radius_of_influence=50000)

    def resample_on_center(self, time: TimeIndexer, longitude: float, latitude: float) -> xr.Dataset:
        da = self.slice_and_stack(time)
        target_def = self._target(latitude, longitude)
        data = self.resample_nearest(da.to_numpy(), target_def)

        # -
        lons, lats = target_def.get_lonlats()
        coords = {LONGITUDE: ((Y_AXIS, X_AXIS), lons), LATITUDE: ((Y_AXIS, X_AXIS), lats), CHANNEL: da.coords[CHANNEL]}
        return self.create_dataset_from_array(data, coords, target_def)

    def slice_and_stack(self, time: TimeIndexer) -> xr.DataArray:
        time_selection = {TIME: time}
        method = self._data.isel if isinstance(time, slice) else self._data.sel
        return method(time_selection).transpose(Y_AXIS, X_AXIS, ...).to_array().stack(channel=(TIME, VARIABLE))

    def create_dataset_from_array(
        self,
        data: np.ndarray,
        coords: Mapping,
        area_definition: pyresample.geometry.AreaDefinition,
        dims=(Y_AXIS, X_AXIS, CHANNEL),
    ) -> xr.Dataset:
        return (
            xr.DataArray(data, dims=dims, coords=coords, attrs={"area_definition": area_definition})
            .unstack(CHANNEL)
            .to_dataset(VARIABLE)
            .transpose(TIME, Y_AXIS, X_AXIS)
        )

    def _repr_html_(self) -> str:
        html = self._data._repr_html_()
        return f"<div><div>{html}</div><div><pre>{repr(self)}</pre></div></div>"

    def __repr__(self) -> str:
        return f"{self._md}\n{self._target}"

    __str__ = __repr__


sampler = GriddedDataSampler(
    ds,
    [TOTAL_CLOUD_COVER, VISIBILITY, CEILING, U_WIND_COMPONENT_10M, V_WIND_COMPONENT_10M, TEMPERATURE_2M],
    target=PartialTargetArea.create("mesoscale-beta", 32, units="km", target_grid="lambert_azimuthal_equal_area"),
)
sampler
```
